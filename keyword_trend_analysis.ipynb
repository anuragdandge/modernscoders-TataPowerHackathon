```python
filepath = r"./data/videoLists1.csv"
data_df = pd.read_csv(filepath)
```

```python
import ast
```

```python
data_df["comments"] = data_df["comments"].apply(ast.literal_eval)
```

```python
data_df.columns
```

```python

# Initialize columns to store positive and negative scores
data_df['positive_score'] = 0
data_df['negative_score'] = 0
data_df['pos_greater_than_neg'] = 0

# Loop through the 'comments' column to calculate positive and negative scores
for index, data in data_df.iterrows():
    pos, neg = 0, 0
    for comment_data in data['comments']:
        if comment_data["sentiment"] == "Positive":
            pos += 1
        elif comment_data["sentiment"] == "Negative":
            neg += 1
    data_df.at[index, 'positive_score'] = pos
    data_df.at[index, 'negative_score'] = neg
    data_df.at[index, 'pos_greater_than_neg'] = (pos - neg) if pos > neg else 0

# Display the updated DataFrame
data_df
```

```python
data_df = data_df.sort_values(by=["pos_greater_than_neg"], ascending=False)
```

```python
data_df["vtags"] = data_df["vtags"].apply(ast.literal_eval)
```
